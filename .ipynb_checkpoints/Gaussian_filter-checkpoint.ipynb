***
Une implémentation entièrement à la main du filtre gaussien ordre 0 donc de base. Le code pour le filtre gaussien est assez intéressant comme il explique comment discrétiser l'implémentation du filtre sur une matrice et définir la taille de la matrice. En revanche, le code pour la convolution reste assez indigeste et on pourra trouver une fonction déjà prête pour cette phase là. 

https://medium.com/@akumar5/computer-vision-gaussian-filter-from-scratch-b485837b6e09

```python
def convolution(oldimage, kernel):
    #image = Image.fromarray(image, 'RGB')
    image_h = oldimage.shape[0]
    image_w = oldimage.shape[1]
    
    
    kernel_h = kernel.shape[0]
    kernel_w = kernel.shape[1]
    
    # récupérer les dimensions de chaque objet mis en cause dans la convolution
    
    if(len(oldimage.shape) == 3):
        image_pad = np.pad(oldimage, pad_width=((kernel_h // 2, kernel_h // 2),(kernel_w // 2, kernel_w // 2),(0,0)), mode='constant',constant_values=0).astype(np.float32)
    elif(len(oldimage.shape) == 2):
        image_pad = np.pad(oldimage, pad_width=((kernel_h // 2, kernel_h // 2),(kernel_w // 2, kernel_w // 2)), mode='constant', constant_values=0).astype(np.float32)
    
    
    h = kernel_h // 2
    w = kernel_w // 2
    
    image_conv = np.zeros(image_pad.shape)
    
    for i in range(h, image_pad.shape[0]-h):
        for j in range(w, image_pad.shape[1]-w):
            #sum = 0
            x = image_pad[i-h:i-h+kernel_h, j-w:j-w+kernel_w]
            x = x.flatten()*kernel.flatten()
            image_conv[i][j] = x.sum()
    h_end = -h
    w_end = -w
    
    if(h == 0):
        return image_conv[h:,w:w_end]
    if(w == 0):
        return image_conv[h:h_end,w:]
    return image_conv[h:h_end,w:w_end]
```

```python
def GaussianBlurImage(image, sigma):
    #image = imread(image)
    # image = Image.open(image)
    image = np.asarray(image)
    #print(image)
    filter_size = 2 * int(4 * sigma + 0.5) + 1 # taille du filtre en fonction du sigma recherché 
    gaussian_filter = np.zeros((filter_size, filter_size), np.float32) # création du filtre vide 
    m = filter_size//2
    n = filter_size//2
    
    for x in range(-m, m+1):
        for y in range(-n, n+1):
            # le filtre est balayé de par le centre d'où la range selon x et selon y 
            x1 = 2*np.pi*(sigma**2)
            x2 = np.exp(-(x**2 + y**2)/(2* sigma**2))
            gaussian_filter[x+m, y+n] = (1/x1)*x2
            # calcul et implémentation en deux temps de chaque valeur du filtre 

    im_filtered = np.zeros_like(image, dtype=np.float32) # crée une copie vide de l'image 
    for c in range(3):
        im_filtered[:, :, c] = convolution(image[:, :, c], gaussian_filter) 
        # convolution image - kernel sur les trois dimension de l'image
    return (im_filtered.astype(np.uint8))
```

```python
img = cv2.imread('test.jpeg')
```

```python
filt = GaussianBlurImage(img, 1.0)
```

```python
plt.figure(figsize=(16, 8)) 
plt.subplot(121),
plt.imshow(img)#,cmap = 'gray')
plt.title('BGR Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),
plt.imshow(filt,cmap = 'gray')
plt.title('Blurred BGR Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

***
L'implémentation précédente à la main est simplifiée avec l'utilisation de la fonction de convolution 2D. Le code précédent est utile pour implémenter facilement tout type de kernel et le code de convolution de la bibliothèque scipy calcule rapidement et de façon compacte la convolution nécessaire pour le filtre. 

```python
sigma=1.0

filter_size = 2 * int(4 * sigma + 0.5) + 1 # taille du filtre en fonction du sigma recherché 
gaussian_filter = np.zeros((filter_size, filter_size), np.float32) # création du filtre vide 
m = filter_size//2
n = filter_size//2
    
for x in range(-m, m+1):
    for y in range(-n, n+1):
        # le filtre est balayé de par le centre d'où la range selon x et selon y 
        x1 = 2*np.pi*(sigma**2)
        x2 = np.exp(-(x**2 + y**2)/(2* sigma**2))
        gaussian_filter[x+m, y+n] = (1/x1)*x2
```

```python
from scipy import signal
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
grad = signal.convolve2d(gray, gaussian_filter, boundary='symm', mode='same')
```

```python
plt.figure(figsize=(16, 8)) 
plt.subplot(121),
plt.imshow(img)#,cmap = 'gray')
plt.title('BGR Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),
plt.imshow(grad,cmap = 'gray')
plt.title('Blurred BGR Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

Visualisation des filtres gaussiens d'ordre 1

```python
sigma=10.0
# theta=np.pi/4
theta=np.pi/3

filter_size = 2 * int(4 * sigma + 0.5) + 1 # taille du filtre en fonction du sigma recherché 
Gx = np.zeros((filter_size, filter_size), np.float32) # création du filtre vide 
Gy = np.zeros((filter_size, filter_size), np.float32) # création du filtre vide
Gcomb = np.zeros((filter_size, filter_size), np.float32) # création du filtre vide
m = filter_size//2
n = filter_size//2
    
for x in range(-m, m+1):
    for y in range(-n, n+1):
        # le filtre est balayé de par le centre d'où la range selon x et selon y 
        Gx[x+m, y+n] = -2*x*np.exp(-(x**2 + y**2)/sigma**2) ## Gx
        Gy[x+m, y+n] = -2*y*np.exp(-(x**2 + y**2)/sigma**2) ## Gy
        Gcomb[x+m, y+n] = np.cos(theta)*Gx[x+m, y+n] + np.sin(theta)*Gy[x+m, y+n]

plt.figure(figsize=(16, 8)) 
plt.subplot(131),
plt.imshow(Gx, cmap='gray')
plt.title('Gx'), plt.xticks([]), plt.yticks([])
plt.subplot(132),
plt.imshow(Gy, cmap='gray')
plt.title('Gy'), plt.xticks([]), plt.yticks([])
plt.subplot(133),
plt.imshow(Gcomb, cmap='gray')
plt.title('Gcomb'), plt.xticks([]), plt.yticks([])

plt.show()
```

Visualisation des filtre gaussiens de 2e ordre

```python
sigma=1.0
# theta=np.pi/4
theta=-np.pi/2

filter_size = 2 * int(4 * sigma + 0.5) + 1 # taille du filtre en fonction du sigma recherché 
Gxx = np.zeros((filter_size, filter_size), np.float32) # création du filtre vide 
Gyy = np.zeros((filter_size, filter_size), np.float32) # création du filtre vide
Gxy = np.zeros((filter_size, filter_size), np.float32) # création du filtre vide
Gcomb = np.zeros((filter_size, filter_size), np.float32) # création du filtre vide
m = filter_size//2
n = filter_size//2
    
for x in range(-m, m+1):
    for y in range(-n, n+1):
        # le filtre est balayé de par le centre d'où la range selon x et selon y 
        Gxx[x+m, y+n] = (4*x**2/sigma**2-2/sigma**2)*np.exp(-(x**2 + y**2)/sigma**2) ## Gxx
        Gyy[x+m, y+n] = (4*y**2/sigma**2-2/sigma**2)*np.exp(-(x**2 + y**2)/sigma**2) ## Gyy
        Gxy[x+m, y+n] = (4*x*y/sigma**2)*np.exp(-(x**2 + y**2)/sigma**2) ## Gxy
        
        Gcomb[x+m, y+n] = np.cos(theta)**2*Gxx[x+m, y+n] + np.sin(theta)**2*Gyy[x+m, y+n] - 2*np.sin(theta)*np.cos(theta)*Gxy[x+m, y+n]
        
plt.figure(figsize=(18, 8)) 
plt.subplot(141),
plt.imshow(Gxx, cmap='gray')
plt.title('Gxx'), plt.xticks([]), plt.yticks([])
plt.subplot(142),
plt.imshow(Gyy, cmap='gray')
plt.title('Gyy'), plt.xticks([]), plt.yticks([])
plt.subplot(143),
plt.imshow(Gxy, cmap='gray')
plt.title('Gxy'), plt.xticks([]), plt.yticks([])
plt.subplot(144),
plt.imshow(Gcomb, cmap='gray')
plt.title('Gcomb'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
import cv2
img = cv2.imread('HL.jpg')
        
from scipy import signal
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
grad = signal.convolve2d(gray, Gcomb, mode='full', boundary='symm')

plt.figure(figsize=(16, 8)) 
plt.subplot(121),
plt.imshow(img)#,cmap = 'gray')
plt.title('BGR Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),
plt.imshow(grad,cmap = 'gray')
plt.title('Blurred BGR Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
print('Grad Size = {}'.format(np.shape(grad)))
print('Max of Grad = {}'.format(np.amax(grad)))
print('Min of Grad = {}'.format(np.amin(grad)))
```

```python
rangee = np.amax(grad) - np.amin(grad)
grad = (grad - np.amin(grad)) * 255 / rangee
plt.figure(figsize=(8,8))
plt.imshow(grad, cmap='gray')
plt.title('Reranged filtered image'), plt.xticks([]), plt.yticks([])

plt.show()
print('Grad Size = {}'.format(np.shape(grad)))
print('Max of Grad = {}'.format(np.amax(grad)))
print('Min of Grad = {}'.format(np.amin(grad)))
```

```python
import cv2
import imutils
import numpy as np
import joblib
 
pts = [] # for storing points
 
 
 # :mouse callback function
def draw_roi(event, x, y, flags, param):
    img2 = img.copy()
 
    if event == cv2.EVENT_LBUTTONDOWN: # Left click, select point
        pts.append((x, y))  
 
    if event == cv2.EVENT_RBUTTONDOWN: # Right click to cancel the last selected point
        pts.pop()  
 
    if event == cv2.EVENT_MBUTTONDOWN: # 
        mask = np.zeros(img.shape, np.uint8)
        points = np.array(pts, np.int32)
        points = points.reshape((-1, 1, 2))
                 # 
        mask = cv2.polylines(mask, [points], True, (255, 255, 255), 2)
        mask2 = cv2.fillPoly(mask.copy(), [points], (255, 255, 255)) # for ROI
        mask3 = cv2.fillPoly(mask.copy(), [points], (0, 255, 0)) # for displaying images on the desktop
 
        show_image = cv2.addWeighted(src1=img, alpha=0.8, src2=mask3, beta=0.2, gamma=0)
 
        cv2.imshow("mask", mask2)
        cv2.imshow("show_img", show_image)
 
        ROI = cv2.bitwise_and(mask2, img)
        cv2.imshow("ROI", ROI)
        cv2.waitKey(0)
 
    if len(pts) > 0:
                 # Draw the last point in pts
        cv2.circle(img2, pts[-1], 3, (0, 0, 255), -1)
 
    if len(pts) > 1:
                 # 
        for i in range(len(pts) - 1):
            cv2.circle(img2, pts[i], 5, (0, 0, 255), -1) # x ,y is the coordinates of the mouse click place
            cv2.line(img=img2, pt1=pts[i], pt2=pts[i + 1], color=(255, 0, 0), thickness=2)
 
    cv2.imshow('image', img2)
 
 
 #Create images and windows and bind windows to callback functions
img = cv2.imread('HL.jpg')
img = imutils.resize(img, width=500)
cv2.namedWindow('image')
cv2.setMouseCallback('image', draw_roi)
print("[INFO] Click the left button: select the point, right click: delete the last selected point, click the middle button: determine the ROI area")
print("[INFO] Press ‘S’ to determine the selection area and save it")
print("[INFO] Press ESC to quit")
while True:
    key = cv2.waitKey(1) & 0xFF
    if key == 27:
        break
    if key == ord("s"):
        saved_data = pts
        #joblib.dump(value=saved_data, filename="config.pkl")
        #print("[INFO] ROI coordinates have been saved to local.")
        print(saved_data)
        break
cv2.destroyAllWindows()

# source https://www.programmersought.com/article/3449903953/
```
